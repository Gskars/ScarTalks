```python
# Definir configurações de geração e segurança
generation_config = {
    "candidate_count": 1,
    "temperature": 0.5,
}

safety_settings = [
    {'category': 'HATE', 'threshold': 'BLOCK_NONE'},
    {'category': 'HARASSMENT', 'threshold': 'BLOCK_NONE'},
    {'category': 'SEXUAL', 'threshold': 'BLOCK_NONE'},
    {'category': 'DANGEROUS', 'threshold': 'BLOCK_NONE'}
]
```

```python
model = genai.GenerativeModel(
    model_name='gemini-1.5-pro',
    generation_config=generation_config,
    safety_settings=safety_settings,
)
```

```python
chat = model.start_chat(history=[])
```

```python
# Função para escolher idioma
def escolher_idioma():
    idiomas = [
        "Inglês", "Francês", "Alemão", "Russo", "Italiano", "Islandês", "Espanhol", "Mandarim", "Romêno","Japonês"
    ]
    print("Escolha o idioma que você quer estudar:")
    for i, idioma in enumerate(idiomas, 1):
        print(f"{i}. {idioma}")

    while True:
        escolha = input("\nDigite o número do idioma desejado: ").strip()
        if escolha.isdigit() and 1 <= int(escolha) <= len(idiomas):
            return idiomas[int(escolha) - 1]
        print("Opção inválida! Por favor, escolha um número de 1 a 10.")

def get_phrasal_verb():
    while True:
        phrasal_verb = input(
            "Digite um phrasal verb (ex: 'look up', 'break down', 'get along'): "
        ).strip()
        if phrasal_verb:
            return phrasal_verb
        print("Por favor, insira um phrasal verb válido.\n")

def gerar_prompt_exemplos(phrasal_verb, tipo, idioma):
    if tipo == "frases":
        prompt = f"""
        Você é um professor de {idioma} especialista em phrasal verbs. Formate toda a resposta em MARKDOWN.
        Siga estas regras rigorosamente:

        **Estrutura da Resposta:**
        1. Título nível 2: **Exemplos com "{phrasal_verb}"**
        2. Primeiro parágrafo: explicação do significado (1 linha)
        3. Lista numerada de 10 itens com:
           - Frase em {idioma} em *itálico*
           - Tradução entre parênteses
           - Exemplo:
             *1. "Look up the address." (Procure o endereço.)*

        **Regras:**
        - Use contextos variados: cotidiano, profissional, viagens
        - Mantenha as traduções simples e literais
        - Não use formatação extra além do especificado
        """
        return prompt

    elif tipo == "dialogos":
        prompt = f"""
        Você é um professor de {idioma} especialista em phrasal verbs. Formate toda a resposta em MARKDOWN.
        Siga estas regras rigorosamente:

        **Estrutura da Resposta:**
        1. Título nível 2: **Diálogos com "{phrasal_verb}"**
        2. Primeiro parágrafo: explicação breve do significado
        3. Para cada um dos 5 diálogos:
           - Subtítulo: **Diálogo [Número] (Contexto)**
           - Personagens fictícios (ex: Alice, Bob, Carol)
           - 2-4 falas por diálogo
           - Cada fala deve conter o phrasal verb
           - Tradução entre parênteses após cada fala

        **Exemplo de Formato:**
        * Alice: Can you look up the recipe? (Você pode procurar a receita?)
        * Bob: I'll look it up online now. (Vou procurar online agora.)

        **Contextos Sugeridos:**
        - Cotidiano - Trabalho - Viagens - Educação - Tecnologia
        """
        return prompt

def gerar_exemplos_outro_idioma(opcao, idioma):
    assunto = input("Qual assunto você quer aprender? ").strip()

    if opcao == "1":
        print(f"\nGerando 10 frases sobre {assunto} em {idioma}...")
        prompt = gerar_prompt_exemplos(assunto, "frases", idioma)
        response = chat.send_message(prompt)  # Substituir pela sua API de IA
        print(f"\n{response.text}\n")

    elif opcao == "2":
        print(f"\nGerando 5 diálogos sobre {assunto} em {idioma}...")
        prompt = gerar_prompt_exemplos(assunto, "dialogos", idioma)
        response = chat.send_message(prompt)  # Substituir pela sua API de IA
        print(f"\n{response.text}\n")

def gerar_frases_e_dialogos(idioma, opcao):
    if idioma == "Inglês":
        phrasal_verb = get_phrasal_verb()
        if opcao == "1":
            prompt = gerar_prompt_exemplos(phrasal_verb, "frases", idioma)
            response = chat.send_message(prompt)  # Substituir pela sua API de IA
            print(f"\n{response.text}\n")

        elif opcao == "2":
            prompt = gerar_prompt_exemplos(phrasal_verb, "dialogos", idioma)
            response = chat.send_message(prompt)  # Substituir pela sua API de IA
            print(f"\n{response.text}\n")

    else:
        gerar_exemplos_outro_idioma(opcao, idioma)

# Main Loop
if __name__ == "__main__":
    while True:
        idioma_escolhido = escolher_idioma()
        print(f"\nVocê escolheu aprender {idioma_escolhido}!\n")

        print("\nEscolha uma opção para gerar exemplos:")
        print("1. Gerar 10 frases sobre o assunto escolhido")
        print("2. Gerar 5 diálogos sobre o assunto escolhido")
        print("3. Sair do chatbot")

        opcao = input("\nDigite a opção desejada: ").strip()

        if opcao == "1" or opcao == "2":
            gerar_frases_e_dialogos(idioma_escolhido, opcao)
        elif opcao == "3":
            print("\nChatbot finalizado. Bons estudos!")
            break
        else:
            print("\nOpção inválida! Por favor, escolha 1, 2 ou 3.\n")
```
